## 1.5. Extracting features

Temporal data is not so useful for classification. Goal is to model features using spectrum of audio file.

### 1.5.1. MFCC Detla and Delta Delta Example
Popular and efficient way of extracting features from audio files.<br>
Report in attachment is detaily explaining how MFCC works.<br>
For now it is enough to know that from every audio file, MFCCs will have shape: [n,m]. Where m is number of frames in which audio is divided and n is number of MFCC coefficient from every frame.<br>
Additional features: delta, delta delta. Same matrix dimensions as MFCC.



```python
SAMPLE_RATE = 44100
MAX_NO_OF_SAMPLES = 1323000
NO_MFCC = 12
```

```python
def featureExtractor(file, pad):
    #opening file
    sound, sr = librosa.core.load(file, sr=SAMPLE_RATE)
    
    if(pad==1):
        if len(sound) < MAX_NO_OF_SAMPLES:
            #print('Filling with zeros...')
            sound = np.pad(sound, (0,MAX_NO_OF_SAMPLES-len(sound)), 'constant')

    #calculating mfcc, delta and delta delta
    mfcc = librosa.feature.mfcc(sound, sr=sr, n_mfcc=NO_MFCC)
    delta = librosa.feature.delta(mfcc, order=1)
    deltadelta = librosa.feature.delta(mfcc, order=2)
    
    return sound, mfcc, delta, deltadelta
```

```python
fname1 = 'audio_train' + '/' + train['fname'][0]

sound1, mfcc1, delta1, deltadelta1 = featureExtractor(fname1,0)

print('sound:', sound1.shape)
print('mfcc, delta, delta delta:', mfcc1.shape)

plt.figure(figsize=(20,10))
plt.subplot(2,2,1),librosa.display.specshow(mfcc1), plt.title('mfcc')
plt.subplot(2,2,2), librosa.display.specshow(delta1), plt.title('delta')
plt.subplot(2,2,3), librosa.display.specshow(deltadelta1), plt.title('deltadelta')
plt.subplot(2,2,4), librosa.display.waveplot(sound1), plt.title('audio time signal')
plt.show()
```

### 1.5.2. MFCC for all files

##### Calculating mfcc, delta and deltadelta as features. Storing whole data.

```python
def mfccDeltaD(dataframe, path):

    XMFCC = np.empty(shape=(dataframe.shape[0], NO_MFCC*MAX_NO_FRAMES))
    XDelta = np.empty(shape=(dataframe.shape[0], NO_MFCC*MAX_NO_FRAMES))
    XDeltaDelta = np.empty(shape=(dataframe.shape[0], NO_MFCC*MAX_NO_FRAMES))
    
    i = 0

    for f in dataframe['fname']:
        
        f = path + '/' + f 
        
        _, mfcc, delta, deltadelta = featureExtractor(f, 1)
        
        XMFCC[i,:] = np.ndarray.flatten(mfcc)
        
        XDelta[i,:] = np.ndarray.flatten(delta)
        
        XDeltaDelta[i,:] = np.ndarray.flatten(deltadelta)
        
        i = i + 1
        
    return XMFCC, XDelta, XDeltaDelta
```

#### Normalization

```python
def normalize(X):

    mean = np.mean(X,axis=0)
    
    std = np.std(X,axis=0)
    std[std==0] = 1

    X = (X - mean) / std
    
    return X
```

#### PCA for matrix

```python
def pcaX(X, ncomp):
    
    pca = PCA(n_components=ncomp)
    
    noExamples = X.shape[0]
    Xc = np.empty(shape=(noExamples, NO_MFCC*ncomp))
    
    for i in range(noExamples):
        
        feat = X[i,:]
        feat = np.reshape(feat, (NO_MFCC, MAX_NO_FRAMES))
        pca.fit(feat.T)
        Xc[i,:] = np.ndarray.flatten(pca.components_)
        
    return Xc
```

#### Average, standard deviation and skewness min and max

```python
def statparams(X):
    
    noExamples = X.shape[0]
    Xc = np.empty(shape=(noExamples, NO_MFCC*5))
    
    for i in range(0, noExamples):
        
        feat = X[i,:]
        feat = np.reshape(feat, (NO_MFCC, MAX_NO_FRAMES))

        avgX = np.mean(feat,axis=1)
        stdX = np.std(feat,axis=1)
        minX = np.min(feat, axis=1)
        maxX = np.max(feat,axis=1)
        skewX = skew(feat, axis=1)

        Xc[i,:] = np.concatenate([avgX, stdX, minX, maxX, skewX])
        
    return Xc
```

#### All features

```python
NO_DATA_TRAIN = 3000
NO_DATA_TEST = 9400

dataTrain = train[train['manually_verified'] == 1].head(n=NO_DATA_TRAIN)
dataTest = test.head(NO_DATA_TEST)
```

```python
start = time.time()

XMFCC, XDelta, XDeltaDelta = mfccDeltaD(dataTrain, 'audio_train')

end = time.time()
print(end-start)
```

```python
start = time.time()

XMFCCTest, XDeltaTest, XDeltaDeltaTest = mfccDeltaD(dataTest, 'audio_test')

end = time.time()
print(end-start)
```

```python
def featureCollector(XMFCC, XDelta, XDeltaDelta, T):
    
        #norm + pca
        if T == 1:
            
            start = time.time()
   
            XMFCCNorm = normalize(XMFCC)
            XDeltaNorm = normalize(XDelta)
            XDeltaDeltaNorm = normalize(XDeltaDelta)

            XMFCCNormPCA = pcaX(XMFCCNorm,5)
            XDeltaNormPCA = pcaX(XDeltaNorm,5)
            XDeltaDeltaNormPCA = pcaX(XDeltaDeltaNorm,5)

            X = np.concatenate([XMFCCNormPCA, XDeltaNormPCA, XDeltaDeltaNormPCA],axis=1)
            print(X.shape)
            end = time.time()
            print(end-start)
            
            return X
           
        # stat moments
        if T == 2:
            
            start = time.time()
            
            XMFCCparam = statparams(XMFCC)
            XDeltaparam = statparams(XDelta)
            XDeltaDeltaparam = statparams(XDeltaDelta)

            X = np.concatenate([XMFCCparam, XDeltaparam, XDeltaDeltaparam],axis=1)
            
            print(X.shape)
            end = time.time()
            print(end-start)
            
            return X
```

#### New features using PCA

```python
#train set:
X1 = featureCollector(XMFCC, XDelta, XDeltaDelta, 1)

#test set:
XTest1 = featureCollector(XMFCCTest, XDeltaTest, XDeltaDeltaTest, 1)
```

#### New features using statisical moments

```python
#train set:
X2 = featureCollector(XMFCC, XDelta, XDeltaDelta, 2)

#test set:
XTest2 = featureCollector(XMFCCTest, XDeltaTest, XDeltaDeltaTest, 2)
```

### 1.5.3. Preparing class 

```python
def labelStrToNum(dataframe):
    
    unq = dataframe['label'].unique()
    num = list(range(0,len(unq)))
    pairs = dict(zip(unq, num))
    pairsInv = dict(zip(num, unq)) #for decoding
    
    y = np.empty(dataframe.shape[0])
    i=0
    
    for label in dataframe['label']:
        y[i] = pairs[label]
        i = i + 1
        
    labels, cnt = np.unique(y, return_counts=True)
    print(dict(zip(labels, cnt)))
    return y, pairsInv
```

```python
y, pairs = labelStrToNum(dataTrain)
print(len(y))
print(pairs)
```

### 1.6.1. KNN

#### Finding best hyper-parameters and dataset

```python
def trainValidateMetric(X, y):
    metrics = ['cosine', 'manhattan', 'minkowski', 'euclidean', 'chebyshev']
    cvsMet = []

    for n in metrics:
        knn = KNeighborsClassifier(metric=n)
        cvsMet.append(np.mean(CVs(knn, X, y, cv=5)))

    plt.figure(figsize=(15,7))
    plt.plot(metrics,cvsMet,'p')
    plt.show()
    
def trainValidateNN(X, y, m):
    nn = [1, 4, 8, 16, 20]
    cvsNn = []

    for n in nn:
        knn = KNeighborsClassifier(n_neighbors=n, metric=m)
        cvsNn.append(np.mean(CVs(knn, X, y, cv=5)))

    plt.figure(figsize=(15,7))
    plt.plot(nn,cvsNn,'p')
    plt.show()
```

PCA data

```python
trainValidateMetric(X1, y)
```

```python
trainValidateNN(X1, y, 'manhattan')
```

Stat moment data

```python
trainValidateMetric(X2, y)
```

```python
trainValidateNN(X2, y, 'manhattan')
```

Best metric is: manhattan. Best number of nearest neighbours: 3, reason for 3 not 1 is possibilty of overfitting. Best train set: Stat moments

Storing results

```python
def store(dataframe, ypred, name):
    clsfpred = pd.DataFrame()
    clsfpred['fname'] = dataframe['fname']
    clsfpred['class'] = ypred
    clsfpred['class'] = clsfpred['class'].apply(lambda x: pairs[x])
    clsfpred.to_csv(name, sep=',', index=False, header=['fname','label'])
```
